```python
# Standard imports
import copy
import seaborn as sns
import numpy as np
import pandas as pd
import xarray as xr
from itertools import product
from multiprocessing import Pool
from simulation import *
from model import *
from parsingfunctions import *
from matplotlib import rcParams
import matplotlib
import matplotlib.pyplot as plt
from cycler import cycler

# Function to configure matplotlib settings
def configure_matplotlib_styles():
    rcParams['font.weight'] = 'bold'
    matplotlib.rc('lines', **{'linewidth': 2, 'mew': 2})
    matplotlib.rc('font', **{'family': 'serif', 'serif': ['Computer Modern'], 'size': 16}, weight='bold')
    matplotlib.rc('text', usetex=True)

# Function to set custom color cycles for plots
def set_color_cycles():
    color_scheme_main = 'green'
    initial_colors = plt.rcParams['axes.prop_cycle'].by_key()['color'][:5]
    custom_marker_cycle = (cycler('linestyle', ["None"]) *
                           cycler('marker', ['x', 'o', 'd', '+', 'P']) +
                           cycler('color', [color_scheme_main] + initial_colors[1:]))
    plt.rc('axes', prop_cycle=custom_marker_cycle)

# Function to set monochrome properties for plots
def set_monochrome_cycles():
    base_color_list = plt.rcParams['axes.prop_cycle'].by_key()['color'][:5]
    line_styles = cycler('linestyle', ['solid', 'dashed', '-.', 'dotted', (0, (3, 5, 1, 5, 1, 5))]) + cycler('color', base_color_list)
    plt.gca().set_prop_cycle(line_styles)
    custom_marker_cycle = cycler('linestyle', ["None"]) * cycler('marker', ['x', 'o', 'd', '+', 'P']) + cycler('color', base_color_list)
    plt.gca().set_prop_cycle(custom_marker_cycle)

# Main function to apply all configurations
def main():
    configure_matplotlib_styles()
    set_color_cycles()
    set_monochrome_cycles()

# Run the main configuration function
if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt
import xarray as xr
from matplotlib.font_manager import FontProperties
import matplotlib

def generate_validation_parameters():
    param_set = {
        'num_devices': list(range(1, 11, 2)),
        'legacy_portion': [0],
        'packet_size': [1000],
        'bandwidth': [20],
        'num_antennas': [256],
        'min_cw': [15, 127, 1023],
        'modulation_scheme': [5],
        'download_type': ['su'],
        'upload_type': ['su'],
        'max_txop_time': [4800],
        'ack_sequence': ['AGGR-MU-BAR'],
        'sched_method': ['rr']
    }
    return param_set

def compute_efficiency(metric_data, max_possible_throughput):
    return (metric_data / max_possible_throughput) * 100

def process_results_for_efficiency(result_data, max_throughput):
    filtered_data = result_data.sel(metrics=['dl', 'ul']).sum('metrics').squeeze()
    return compute_efficiency(filtered_data, max_throughput)

def plot_throughput_vs_stations(model_data, sim_data, config_params, plot_title, output_file):
    plt.figure(figsize=(10, 6))

    for idx, current_cw in enumerate(config_params['min_cw']):
        model_data.sel(min_cw=current_cw).plot.line(
            x='num_devices', linestyle='-', linewidth=2, markersize=8, 
            label=fr'\textbf{{Model, CWmin = {current_cw}}}'
        )

    for idx, current_cw in enumerate(config_params['min_cw']):
        sim_data.sel(min_cw=current_cw).plot.line(
            x='num_devices', linestyle='--', linewidth=2, markersize=8, 
            label=fr'\textbf{{Simulation, CWmin = {current_cw}}}'
        )

    plt.xlabel(r'\textbf{Number of Devices}')
    plt.ylabel(r'\textbf{Throughput [Mbit/s]}')
    plt.xticks(config_params['num_devices'])

    plt.plot(config_params['num_devices'], processed_model_efficiency, 
             label=r'\textbf{Model Efficiency}', linestyle=':', marker='o')
    plt.plot(config_params['num_devices'], processed_sim_efficiency, 
             label=r'\textbf{Simulation Efficiency}', linestyle=':', marker='s')

    legend_items, legend_texts = plt.gca().get_legend_handles_labels()

    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=True, prop={'weight': 'bold'})
    plt.grid(True, axis='both', which='both')
    plt.title(plot_title)

    plt.savefig(output_file, bbox_inches='tight', pad_inches=0.5, dpi=500)
    plt.show()

validation_params = generate_validation_parameters()
model_throughput = get_model_throughput(validation_params)

sim_repetitions = 2
simulation_params = adapt_dictionary_for_sims(validation_params, step=2, simulationTime=5)
simulation_metrics = get_simulation_metrics(simulation_params, runs=sim_repetitions, overwrite=False, results_folder='ofdma-validation-results')

max_throughput_value = 100

processed_model_efficiency = process_results_for_efficiency(model_throughput, max_throughput_value)

if 'runs' in simulation_metrics.dims:
    averaged_sim_results = simulation_metrics.mean('runs')
    sim_throughput = averaged_sim_results.sel(metrics=['dl', 'ul']).sum('metrics').squeeze()
    processed_sim_efficiency = compute_efficiency(sim_throughput, max_throughput_value)
else:
    sim_throughput = simulation_metrics.sel(metrics=['dl', 'ul']).sum('metrics').squeeze()
    processed_sim_efficiency = compute_efficiency(sim_throughput, max_throughput_value)

plot_throughput_vs_stations(
    model_data=model_throughput,
    sim_data=sim_throughput,
    config_params=validation_params,
    plot_title=r'\textbf{Aggregate Throughput vs Number of Devices}',
    output_file='figures/specialFourRevised.png'
)
```

```python
import matplotlib.pyplot as plt
import xarray as xr

def compute_improvement(throughput_data, max_throughput_value):
    return (throughput_data / max_throughput_value) * 100

def generate_validation_params():
    config_parameters = {
        'num_stations': list(range(1, 10)),
        'legacy_fraction': [0],
        'frame_size': [1000],
        'channel_bandwidth': [20],
        'num_antennas': [256],
        'min_cw_sizes': [15],
        'modulation_scheme': [5],
        'downlink_type': ['mu'],
        'uplink_type': ['None'],
        'max_txop_duration': [4800],
        'ack_sequence_type': ['AGGR-MU-BAR'],
        'scheduling_method': ['rr', 'bellalta']
    }
    return config_parameters

def plot_throughput_results(model_data, sim_data, station_counts, model_improvement, sim_improvement):
    plt.figure(figsize=(10, 6))
    
    model_data.sel(metrics='hol').squeeze().plot.line(
        x='num_stations', linestyle='-', linewidth=2, label=r'\textbf{Model Bianchi Throughput}'
    )
    
    sim_data.sel(metrics='hol').mean('runs').squeeze().plot.line(
        x='num_stations', linestyle='--', linewidth=2, label=r'\textbf{Simulation Throughput}'
    )
    
    plt.plot(station_counts, sim_improvement, label=r'\textbf{Simulation Improvement}', linestyle=':', marker='s')
    plt.plot(station_counts, model_improvement, label=r'\textbf{Model Bianchi Improvement}', linestyle=':', marker='o')
    
    plt.title(r'\textbf{HoL Delay for Different Numbers of STAs with Contention}')
    plt.xlabel(r'\textbf{Number of STAs}')
    plt.ylabel(r'\textbf{Head-of-Line Delay [ms]}')
    plt.xticks(station_counts)
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=True)
    plt.grid(True, axis='both', which='both')
    plt.savefig('figures/special_Two_Revised.png', bbox_inches='tight', dpi=500)
    plt.show()

def main():
    validation_params = generate_validation_params()
    
    sim_run_count = 2
    adapted_sim_params = adapt_dictionary_for_sims(validation_params, step=1, simulationTime=5)
    
    simulation_results = get_simulation_metrics(
        adapted_sim_params, runs=sim_run_count, overwrite=False, results_folder='ofdma-validation-results'
    )
    
    model_throughput_data = get_model_throughput(validation_params).squeeze()
    max_throughput_placeholder = 100
    
    model_throughput_improvement = compute_improvement(
        model_throughput_data.sel(metrics='hol'), max_throughput_placeholder
    )
    
    simulation_throughput_improvement = compute_improvement(
        simulation_results.sel(metrics='hol').mean('runs').squeeze(), max_throughput_placeholder
    )
    
    plot_throughput_results(
        model_throughput_data,
        simulation_results,
        validation_params['num_stations'],
        model_throughput_improvement,
        simulation_throughput_improvement
    )
    
    extended_sim_params = {
        **validation_params,
        'uplink_type': ['mu'],
        'min_cw_sizes': [15, 31, 63, 127]
    }

if __name__ == "__main__":
    main()
```

```python
import matplotlib.pyplot as plt

def generate_validation_parameters():
    config_params = {
        'num_stations': list(range(1, 10, 3)),
        'legacy_fraction': [0],
        'frame_size': [1000],
        'channel_bandwidth': [20],
        'num_antennas': [256],
        'contention_window_min': [15, 63],
        'modulation_scheme': [5],
        'downlink_traffic': ['mu'],
        'uplink_traffic': ['su'],
        'max_txop_duration': [4800],
        'ack_sequence_type': ['AGGR-MU-BAR'],
        'scheduling_method': ['rr', 'bellalta']
    }
    return config_params

def plot_model_throughput(model_data, contention_levels):
    plt.figure(figsize=(10, 10))
    for contention in contention_levels:
        model_data.sel(metrics='dl', cwMin=contention).squeeze().plot.line(
            x='num_stations', label=fr'\textbf{{Model Bianchi, CWmin = {contention}}}'
        )

def plot_simulation_results(sim_data, contention_levels, max_throughput, sim_station_counts):
    for contention in contention_levels:
        sim_selected_data = sim_data.sel(metrics='dl', cwMin=contention).mean('runs').squeeze()
        sim_selected_data.plot.line(
            x='num_stations', label=fr'\textbf{{Sim, CWmin = {contention}}}'
        )
        
        sim_improvement = (sim_selected_data / max_throughput) * 100
        plt.plot(sim_station_counts, sim_improvement, linestyle='--', marker='x', label=r'\textbf{Improved Sim}')

def main():
    validation_config = generate_validation_parameters()
    
    model_throughput_data = get_model_throughput(validation_config).squeeze()
    plot_model_throughput(model_throughput_data, validation_config['contention_window_min'])
    
    simulation_run_count = 2
    adapted_sim_params = adapt_dictionary_for_sims(validation_config, step=1, simulationTime=5)
    
    sim_results_data = get_simulation_metrics(
        adapted_sim_params, runs=simulation_run_count, overwrite=False, results_folder='ofdma-validation-results'
    )
    
    max_possible_throughput = 100
    plot_simulation_results(sim_results_data, adapted_sim_params['cwMin'], max_possible_throughput, adapted_sim_params['num_stations'])
    
    plt.title(r'\textbf{Aggregate DL Throughput with Different Contention Levels}')
    plt.xlabel(r'\textbf{Number of STAs}')
    plt.ylabel(r'\textbf{Throughput [Mbit/s]}')
    plt.xticks(adapted_sim_params['num_stations'])
    plt.grid(True, axis='both', which='both')
    plt.tight_layout()
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.15), ncol=2, frameon=True)
    plt.savefig('figures/special_one_RevisedC1.png', bbox_inches='tight', dpi=500)
    plt.show()

if __name__ == "__main__":
    main()
```

# Special Ack Edition

```python
import matplotlib.pyplot as plt
from cycler import cycler
from matplotlib.lines import Line2D

def setup_plot_style(new_color, default_colors):
    plt.figure(figsize=(12, 6))
    marker_styles = (cycler('linestyle', ["None"]) *
                     cycler('marker', ['x', 'o', 'd', '+', 'P']) +
                     cycler('color', [new_color] + default_colors[:4]))
    return list(marker_styles), [marker['marker'] for marker in list(marker_styles)]

def compute_improvement(data, max_throughput):
    return (data / max_throughput) * 100

def plot_throughput_data(model_data, sim_data, ack_types, contention_levels, line_styles, marker_styles):
    for ack_idx, ack_type in enumerate(ack_types):
        for cw_idx, cw_value in enumerate(contention_levels):
            model_series = model_data.sel(metrics='dl', dl='mu', ackSeqType=ack_type, cwMin=cw_value).squeeze()
            plt.plot(model_series['nStations'], model_series, 
                     label=f'Model, {ack_type}, CWmin={cw_value}' if cw_idx == 0 else "",
                     linestyle=line_styles[cw_idx], marker=marker_styles[cw_idx])
            
            sim_series = sim_data.sel(metrics='dl', dl='mu', ackSeqType=ack_type, cwMin=cw_value).mean('runs').squeeze()
            plt.plot(sim_series['nStations'], sim_series, 
                     label=f'Sim, {ack_type}, CWmin={cw_value}' if cw_idx == 0 else "",
                     linestyle=line_styles[cw_idx], marker=marker_styles[cw_idx])

def add_legends(ack_types, marker_styles, contention_levels, default_colors):
    ack_legend = [Line2D([0], [0], color=default_colors[i], linestyle='-', 
                         label=fr'\textbf{{Model, {ack_type}}}') for i, ack_type in enumerate(ack_types)]
    contention_legend = [Line2D([0], [0], color='black', marker=marker, linestyle='None', 
                                label=f'CWmin={cw}') for marker, cw in zip(marker_styles, contention_levels)]
    
    plt.subplots_adjust(bottom=0.3)
    plt.legend(handles=ack_legend, loc='upper center', bbox_to_anchor=(0.5, -0.15), 
               fancybox=True, frameon=True, shadow=True, ncol=3, fontsize='small')
    plt.legend(handles=contention_legend, loc='lower center', bbox_to_anchor=(0.5, -0.4), 
               fancybox=True, shadow=True, frameon=True, ncol=3, fontsize='small')

def main():
    config_params = {
        'num_stations': list(range(1, 10, 4)),
        'legacy_fraction': [0],
        'frame_size': [1000],
        'channel_bandwidth': [20],
        'num_antennas': [256],
        'contention_window_min': [15, 255],
        'modulation_scheme': [5],
        'downlink_traffic': ['su', 'mu'],
        'uplink_traffic': ['None'],
        'max_txop_duration': [4800],
        'ack_sequence_type': ['ACK-SU-FORMAT', 'MU-BAR', 'AGGR-MU-BAR'],
        'scheduling_method': ['rr']
    }
    
    new_color_style = '#1f77b4'
    color_palette = plt.rcParams['axes.prop_cycle'].by_key()['color'][:5]
    if '#2ca02c' in color_palette:
        color_palette.remove('#2ca02c')
    
    marker_styles, markers = setup_plot_style(new_color_style, color_palette)
    line_styles = ['solid', 'dashed', '-.', 'dotted', (0, (3, 5, 1, 5, 1, 5))]

    model_throughput = get_model_throughput(config_params)
    theoretical_max_throughput = 100
    
    model_dl_data = model_throughput.sel(metrics='dl', dl='mu').mean('ackSeqType').mean('cwMin').squeeze()
    model_improvement = compute_improvement(model_dl_data, theoretical_max_throughput)

    sim_params = adapt_dictionary_for_sims(config_params, step=4, simulationTime=3)
    simulation_output = get_simulation_metrics(sim_params, runs=2, overwrite=False, results_folder='ofdma-validation-results')
    
    sim_dl_data = simulation_output.sel(metrics='dl', dl='mu').mean('ackSeqType').mean('cwMin').mean('runs').squeeze()
    sim_improvement = compute_improvement(sim_dl_data, theoretical_max_throughput)

    plot_throughput_data(model_throughput, simulation_output, config_params['ack_sequence_type'], 
                         config_params['contention_window_min'], line_styles, markers)

    add_legends(config_params['ack_sequence_type'], markers, config_params['contention_window_min'], color_palette)

    plt.plot(model_dl_data['nStations'], model_improvement, label='Model Improvement', linestyle=':', marker='o')
    plt.plot(sim_dl_data['nStations'], sim_improvement, label='Simulation Improvement', linestyle=':', marker='s')

    plt.title(r'\textbf{DL Throughput for Different ACK Sequences and Contention Levels}')
    plt.xlabel(r'\textbf{Number of STAs}')
    plt.ylabel(r'\textbf{Throughput [Mbit/s]}')
    plt.xticks(range(1, 10, 4))
    plt.grid(True, which='both', axis='both')
    plt.savefig('figures/ack_Special_EditionC1.png', bbox_inches='tight', pad_inches=0.5, dpi=500)
    plt.show()

if __name__ == "__main__":
    main()
```
